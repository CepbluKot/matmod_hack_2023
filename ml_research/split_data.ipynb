```python
# split data by eng type, flight phase, n1 modifier


import pandas as pd
import re


orig_df = pd.read_csv('X.csv')

engine_types = orig_df['engine_type'].unique()
flight_phases = orig_df['flight_phase'].unique()
n1_modifiers = orig_df['n1_modifier'].unique()



# split by eng, phase, n1
for engine_type in engine_types:
    for flight_phase in flight_phases:
        for n1_modifier in n1_modifiers:
            engine_type = re.sub(r'[/]', '', engine_type)
            filename = str(engine_type) + '_' + str(n1_modifier) + '_' + str(flight_phase)

            to_save_df = orig_df.loc[(orig_df['engine_type'] == engine_type) & (orig_df['flight_phase'] == flight_phase) & (orig_df['n1_modifier'] == n1_modifier)]


            if len(to_save_df):
                print(filename ,len(to_save_df))
                to_save_df.to_csv('/home/oleg/Documents/matmod_challeng/splitted_x_data/by_eng_phase_n1/'+filename)



print('-------------')
# split by phase

for flight_phase in flight_phases:
    filename = str(flight_phase)

    to_save_df = orig_df.loc[(orig_df['flight_phase'] == flight_phase)]

    if len(to_save_df):
        print(filename ,len(to_save_df))
        to_save_df.to_csv('/home/oleg/Documents/matmod_challeng/splitted_x_data/by_phase/'+filename)

```
